```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os
os.getcwd()
```

```python
df = pd.read_csv("student_exam_scores.csv.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df["exam_score"], bins="auto")
plt.title("Histogram: Exam Score Distribution")
plt.xlabel("Exam Score")
plt.ylabel("Count")
plt.grid(True, linestyle=':', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["hours_studied"], df["exam_score"], alpha=0.6)
# add trendline (linear fit)
m, b = np.polyfit(df["hours_studied"], df["exam_score"], 1)
plt.plot(df["hours_studied"], m*df["hours_studied"] + b, linestyle='--')
plt.title("Scatter: hours_studied vs exam_score")
plt.xlabel("hours_studied")
plt.ylabel("exam_score")
plt.grid(True, linestyle=':', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["attendance_percent"], df["exam_score"], alpha=0.6)
m2, b2 = np.polyfit(df["attendance_percent"], df["exam_score"], 1)
plt.plot(df["attendance_percent"], m2*df["attendance_percent"] + b2, linestyle='--')
plt.title("Scatter: attendance_percent vs exam_score")
plt.xlabel("attendance_percent")
plt.ylabel("exam_score")
plt.grid(True, linestyle=':', linewidth=0.5)
plt.tight_layout()
plt.savefig("scatter_attendance_final.png")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["sleep_hours"], df["exam_score"], alpha=0.6)
m3, b3 = np.polyfit(df["sleep_hours"], df["exam_score"], 1)
plt.plot(df["sleep_hours"], m3*df["sleep_hours"] + b3, linestyle='--')
plt.title("Scatter: sleep_hours vs exam_score")
plt.xlabel("sleep_hours")
plt.ylabel("exam_score")
plt.grid(True, linestyle=':', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["previous_scores"], df["exam_score"], alpha=0.6)
m4, b4 = np.polyfit(df["previous_scores"], df["exam_score"], 1)
plt.plot(df["previous_scores"], m4*df["previous_scores"] + b4, linestyle='--')
plt.title("Scatter: previous_scores vs exam_score")
plt.xlabel("previous_scores")
plt.ylabel("exam_score")
plt.grid(True, linestyle=':', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
corr = df[["hours_studied","sleep_hours","attendance_percent","previous_scores","exam_score"]].corr()

plt.figure(figsize=(6,5))
plt.imshow(corr, interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        plt.text(j, i, f"{corr.iloc[i,j]:.2f}", ha="center", va="center", color="w" if abs(corr.iloc[i,j])>0.5 else "k")
plt.title("Correlation Matrix (numeric values shown)")
plt.tight_layout()
plt.show()
```
